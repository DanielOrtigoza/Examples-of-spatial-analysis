```python
import osmnx as ox
import folium
```

# Main code

```python
final_map = folium.Map(location = [40.7128, -74.0060], zoom_start = 14)

graph = ox.graph_from_point((40.7128, -74.0060), dist = 1000, network_type = "drive")
nodes = graph.nodes(data = True)

for node, data in nodes:
  folium.Marker(
    location = [data['y'], data['x']],
  ).add_to(final_map)
```
